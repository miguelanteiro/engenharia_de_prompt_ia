```python
# Simula a temperatura do sistema em Celsius. Você pode alterar este valor para testar diferentes cenários.
temperature = 35

# Verifica se a temperatura atual é maior que 30 graus Celsius.
# Este é o limite definido para detectar um possível superaquecimento.
if temperature > 30:
    # Se a temperatura for maior que 30, exibe uma mensagem de alerta.
    # Usa concatenação de strings para incluir o valor da temperatura na mensagem.
    print("ALERTA: Temperatura do sistema é de " + str(temperature) + "°C! Superaquecimento detectado.")
else:
    # Se a temperatura for 30 graus Celsius ou menor, indica que está tudo normal.
    print("Temperatura do sistema é de " + str(temperature) + "°C. Tudo normal.")
    # Adiciona uma linha em branco para melhorar a legibilidade da saída.
    print()
```
