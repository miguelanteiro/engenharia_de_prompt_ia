# Automação de arquivos

Crie um programa que oranize arquivos bagunados em pastas automaticamente, baseando-se apenas a extensão do arquivo (ex: .pdf, .jpg).

### Organização de Arquivos por Extensão

Este script Python irá organizar arquivos em um diretório especificado, movendo-os para subpastas criadas automaticamente com base em suas extensões. Por exemplo, todos os arquivos `.pdf` irão para uma pasta chamada `PDF`.

```python
import os
import shutil

# Defina o diretório que você deseja organizar.
# CUIDADO: Este script moverá arquivos. É recomendável testá-lo em uma cópia de seus arquivos ou em um diretório de teste.
source_directory = 'arquivos_baguncados' # Altere para o caminho do seu diretório

# Verifica se o diretório existe
if not os.path.exists(source_directory):
    print(f"O diretório '{source_directory}' não foi encontrado. Criando um para demonstração...")
    os.makedirs(source_directory)
    # Criar alguns arquivos de exemplo para teste
    open(os.path.join(source_directory, 'documento.pdf'), 'w').close()
    open(os.path.join(source_directory, 'foto.jpg'), 'w').close()
    open(os.path.join(source_directory, 'relatorio.docx'), 'w').close()
    open(os.path.join(source_directory, 'planilha.xlsx'), 'w').close()
    open(os.path.join(source_directory, 'apresentacao.pptx'), 'w').close()
    open(os.path.join(source_directory, 'outro.txt'), 'w').close()
    open(os.path.join(source_directory, 'video.mp4'), 'w').close()
    open(os.path.join(source_directory, 'imagem.png'), 'w').close()
    print("Arquivos de exemplo criados. Execute novamente para organizar.")
else:
    print(f"Organizando arquivos em: {source_directory}")

    # Lista todos os arquivos e diretórios no diretório de origem
    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename)

        # Ignora diretórios
        if os.path.isdir(file_path):
            continue

        # Obtém a extensão do arquivo (ex: '.pdf', '.jpg')
        file_extension = os.path.splitext(filename)[1].lower()

        # Se não houver extensão, move para uma pasta 'OUTROS'
        if not file_extension:
            destination_folder = os.path.join(source_directory, 'OUTROS')
        else:
            # Cria o nome da pasta com base na extensão (ex: '.pdf' -> 'PDF')
            destination_folder = os.path.join(source_directory, file_extension[1:].upper())

        # Cria a pasta de destino se ela não existir
        if not os.path.exists(destination_folder):
            os.makedirs(destination_folder)

        # Move o arquivo para a pasta de destino
        shutil.move(file_path, destination_folder)
        print(f"Movido: {filename} -> {destination_folder}/")

    print("Organização concluída!")
```

### Como Usar:

1.  **Defina o diretório**: No código acima, altere a variável `source_directory` para o caminho da pasta que você deseja organizar.
2.  **Execute**: Execute a célula de código. Se o diretório não existir ou estiver vazio, ele criará um e alguns arquivos de exemplo para que você possa testar.
3.  **Verifique**: Após a execução, seu diretório `source_directory` terá subpastas (ex: `PDF`, `JPG`, `DOCX`) e os arquivos serão movidos para as pastas correspondentes.

# Consulta Simples e APIs

Desenvolva um script que consulte uma API publica (viaCep) e retorne as informações de endereço formatadas de maneira legivel.


```python
import requests

def consultar_cep(cep):
    """Consulta a API ViaCEP para obter informações de endereço."""
    # Remove qualquer caractere não numérico do CEP
    cep = ''.join(filter(str.isdigit, cep))

    if len(cep) != 8:
        return {"erro": True, "mensagem": "CEP inválido. Deve conter 8 dígitos."}

    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status() # Lança um erro para status de resposta HTTP ruins (4xx ou 5xx)
        data = response.json()

        if "erro" in data and data["erro"]:
            return {"erro": True, "mensagem": "CEP não encontrado."}
        else:
            return {"erro": False, "data": data}
    except requests.exceptions.RequestException as e:
        return {"erro": True, "mensagem": f"Erro ao conectar com a API ViaCEP: {e}"}
    except ValueError:
        return {"erro": True, "mensagem": "Resposta inválida da API. Não foi possível decodificar JSON."}

# Exemplo de uso
cep_digitado = input("Digite o CEP para consulta: ")
resultado = consultar_cep(cep_digitado)

if resultado["erro"]:
    print(f"Erro: {resultado['mensagem']}")
else:
    endereco = resultado["data"]
    print("\n--- Informações do Endereço ---")
    print(f"CEP: {endereco.get('cep', 'N/A')}")
    print(f"Logradouro: {endereco.get('logradouro', 'N/A')}")
    print(f"Complemento: {endereco.get('complemento', 'N/A')}")
    print(f"Bairro: {endereco.get('bairro', 'N/A')}")
    print(f"Cidade: {endereco.get('localidade', 'N/A')}")
    print(f"Estado (UF): {endereco.get('uf', 'N/A')}")
    print(f"IBGE: {endereco.get('ibge', 'N/A')}")
    print(f"DDD: {endereco.get('ddd', 'N/A')}")
    print("-------------------------------")
```

# Sistema de Notificação

Crie um programa autonomo que dispare alertas (mensagens de alerta no console ou e-mail simulado) unicamente quando uma condição critica for atingida

### Monitoramento e Alerta de Condição Crítica

Este programa simula um sistema de monitoramento que verifica uma condição (neste caso, a temperatura) e dispara um alerta quando essa condição se torna crítica. O alerta pode ser uma mensagem no console ou, em um cenário real, um e-mail ou SMS.

```python
import time
import random

def send_alert(message, notification_type='console'):
    """Simula o envio de um alerta (console ou e-mail)."""
    if notification_type == 'console':
        print(f"\n!!! ALERTA DE NOTIFICAÇÃO (CONSOLE) !!!")
        print(f"Horário: {time.ctime()}")
        print(f"Mensagem: {message}")
        print(f"!!! FIM DO ALERTA !!!\n")
    elif notification_type == 'email_simulado':
        print(f"\n--- E-MAIL SIMULADO ENVIADO ---")
        print(f"De: sistema@exemplo.com")
        print(f"Para: admin@exemplo.com")
        print(f"Assunto: ALERTA DE CONDIÇÃO CRÍTICA")
        print(f"Horário: {time.ctime()}")
        print(f"Corpo: {message}")
        print(f"------------------------------\n")
    else:
        print("Tipo de notificação inválido.")


def monitor_system(critical_threshold=30, check_interval_seconds=2, notification_type='console', num_checks=5):
    """Monitora uma condição e dispara alertas se a condição for crítica."""
    print(f"Iniciando monitoramento do sistema. Limite crítico: {critical_threshold}°C")
    print(f"Verificando a cada {check_interval_seconds} segundos. Total de {num_checks} verificações.")

    for i in range(num_checks):
        # Simula a leitura da temperatura atual
        # Em um cenário real, isso viria de um sensor ou API
        current_temperature = random.randint(25, 38) # Temperatura aleatória entre 25 e 38°C
        print(f"[{i+1}/{num_checks}] Temperatura atual: {current_temperature}°C")

        if current_temperature > critical_threshold:
            alert_message = f"TEMPERATURA ELEVADA! A temperatura do sistema atingiu {current_temperature}°C, excedendo o limite de {critical_threshold}°C."
            send_alert(alert_message, notification_type)
        else:
            print("Temperatura normal. Tudo ok.")

        time.sleep(check_interval_seconds)
    print("Monitoramento concluído.")
```

```python
# Exemplo de uso do sistema de notificação
# Você pode alterar 'console' para 'email_simulado' para ver o e-mail simulado
# Ou ajustar 'critical_threshold' e 'num_checks'
monitor_system(critical_threshold=30, check_interval_seconds=1, notification_type='console', num_checks=10)
```
