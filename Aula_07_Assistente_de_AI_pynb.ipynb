Calculadora simples


```python
def adicionar(x, y):
    return x + y

def subtrair(x, y):
    return x - y

print("Selecione a operação:")
print("1. Adicionar")
print("2. Subtrair")

while True:
    escolha = input("Digite sua escolha (1/2): ")

    if escolha in ('1', '2'):
        try:
            num1 = float(input("Digite o primeiro número: "))
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")
            continue

        if escolha == '1':
            print(f"{num1} + {num2} = {adicionar(num1, num2)}")

        elif escolha == '2':
            print(f"{num1} - {num2} = {subtrair(num1, num2)}")

        proxima_conta = input("Deseja fazer outra conta? (sim/não): ")
        if proxima_conta.lower() != 'sim':
            break
    else:
        print("Entrada inválida. Por favor, digite 1 ou 2.")
```

### Como usar a calculadora:

1.  **Execute a célula de código** acima.
2.  Você verá um menu para selecionar a operação (Adicionar ou Subtrair).
3.  **Digite `1` para adição** ou **`2` para subtração** e pressione Enter.
4.  Em seguida, **digite o primeiro número** e pressione Enter.
5.  **Digite o segundo número** e pressione Enter.
6.  O resultado será exibido.
7.  Você terá a opção de fazer outra conta. Digite `sim` para continuar ou qualquer outra coisa para sair.
