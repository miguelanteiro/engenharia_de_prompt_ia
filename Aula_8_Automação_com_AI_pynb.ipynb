# Automação de arquivos

Crie um programa que oranize arquivos bagunados em pastas automaticamente, baseando-se apenas a extensão do arquivo (ex: .pdf, .jpg).

### Organização de Arquivos por Extensão

Este script Python irá organizar arquivos em um diretório especificado, movendo-os para subpastas criadas automaticamente com base em suas extensões. Por exemplo, todos os arquivos `.pdf` irão para uma pasta chamada `PDF`.

```python
import os
import shutil

# Defina o diretório que você deseja organizar.
# CUIDADO: Este script moverá arquivos. É recomendável testá-lo em uma cópia de seus arquivos ou em um diretório de teste.
source_directory = 'arquivos_baguncados' # Altere para o caminho do seu diretório

# Verifica se o diretório existe
if not os.path.exists(source_directory):
    print(f"O diretório '{source_directory}' não foi encontrado. Criando um para demonstração...")
    os.makedirs(source_directory)
    # Criar alguns arquivos de exemplo para teste
    open(os.path.join(source_directory, 'documento.pdf'), 'w').close()
    open(os.path.join(source_directory, 'foto.jpg'), 'w').close()
    open(os.path.join(source_directory, 'relatorio.docx'), 'w').close()
    open(os.path.join(source_directory, 'planilha.xlsx'), 'w').close()
    open(os.path.join(source_directory, 'apresentacao.pptx'), 'w').close()
    open(os.path.join(source_directory, 'outro.txt'), 'w').close()
    open(os.path.join(source_directory, 'video.mp4'), 'w').close()
    open(os.path.join(source_directory, 'imagem.png'), 'w').close()
    print("Arquivos de exemplo criados. Execute novamente para organizar.")
else:
    print(f"Organizando arquivos em: {source_directory}")

    # Lista todos os arquivos e diretórios no diretório de origem
    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename)

        # Ignora diretórios
        if os.path.isdir(file_path):
            continue

        # Obtém a extensão do arquivo (ex: '.pdf', '.jpg')
        file_extension = os.path.splitext(filename)[1].lower()

        # Se não houver extensão, move para uma pasta 'OUTROS'
        if not file_extension:
            destination_folder = os.path.join(source_directory, 'OUTROS')
        else:
            # Cria o nome da pasta com base na extensão (ex: '.pdf' -> 'PDF')
            destination_folder = os.path.join(source_directory, file_extension[1:].upper())

        # Cria a pasta de destino se ela não existir
        if not os.path.exists(destination_folder):
            os.makedirs(destination_folder)

        # Move o arquivo para a pasta de destino
        shutil.move(file_path, destination_folder)
        print(f"Movido: {filename} -> {destination_folder}/")

    print("Organização concluída!")
```

### Como Usar:

1.  **Defina o diretório**: No código acima, altere a variável `source_directory` para o caminho da pasta que você deseja organizar.
2.  **Execute**: Execute a célula de código. Se o diretório não existir ou estiver vazio, ele criará um e alguns arquivos de exemplo para que você possa testar.
3.  **Verifique**: Após a execução, seu diretório `source_directory` terá subpastas (ex: `PDF`, `JPG`, `DOCX`) e os arquivos serão movidos para as pastas correspondentes.
